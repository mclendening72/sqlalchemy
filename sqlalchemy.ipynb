```python
import sqlite3

conn = sqlite3.connect("books.db")
cur = conn.cursor()
cur.execute("CREATE TABLE books (title TEXT, author TEXT, year INTEGER)")
conn.commit()
conn.close()
```

```python
import csv
import sqlite3

conn = sqlite3.connect("books.db")
cur = conn.cursor()

with open("books2.csv", newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        cur.execute("INSERT INTO books (title, author, year) VALUES (?, ?, ?)",
                    (row["title"], row["author"], int(row["year"])))

conn.commit()
conn.close()
```

```python
from sqlalchemy import create_engine, text

engine = create_engine("sqlite:///books.db")

with engine.connect() as conn:
    result = conn.execute(text("SELECT title FROM books ORDER BY title"))
    for row in result:
        print(row.title)
```
